```python
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd
import matplotlib.pyplot as plt
import tempfile
from prefect.engine.results import S3Result
from prefect.engine.serializers import Serializer
import prefect

class NoOpSerializer(Serializer):
    """A `Serializer` that does nothing."""

    def serialize(self, value):
        return value

    def deserialize(self, value):
        return value
    
MARKDOWN = """# {coin_name}
<img src="https://omlds-prefect.s3.amazonaws.com/{image}">
"""

df = pd.DataFrame({'a':[1,2,3,4,5,6,7,6,5,4,3,2,1,2,3,4,5,6,7,1]})
model = ARIMA(df['a'], order=(1,1,0))
model_fit = model.fit()
```

```python
# predictions
df['set'] = 'train'
df2 = pd.DataFrame({'a': model_fit.forecast(5)})
df2['set'] = 'forecast'
df = df[['a', 'set']]
df = pd.concat([df, df2], axis=0, ignore_index=True)
```

```python
import seaborn as sns
sns.lineplot(list(range(0,len(df['a']))), df['a'], hue=df['set'])
```
